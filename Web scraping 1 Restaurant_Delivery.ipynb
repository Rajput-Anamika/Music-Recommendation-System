```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.brightlocal.com/resources/top-review-sites/review-sites-by-industry/restaurant-review-sites/'
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify)
```

```python
data=[]
td_review = soup.find_all('td')
for i in td_review:
    data.append((i.get_text()))
```

```python
data
```

```python
# Convert the flat list into a list of sublists with four elements each
n = 4
data_sublists = [data[i:i+n] for i in range(0, len(data), n)]

# Convert the list of sublists into a DataFrame
df = pd.DataFrame(data_sublists, columns=['Review Site', 'Domain Authority', 'Country', 'Free/Paid'])

print(df)
```

```python
df.to_csv('restaurant_delivery.csv', index=False)
```
